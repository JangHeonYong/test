```python

# ---------------------------------------------
# 산업경영공학 진출분야 추천 프로그램 (선택 분야만 입력)
# ---------------------------------------------

#액셀 저장 라이브러리
from openpyxl import Workbook

# 1. 학점 변환
grade_to_score = {
    "A+": 4.5, "A0": 4.0,
    "B+": 3.5, "B0": 3.0,
    "C+": 2.5, "C0": 2.0,
    "D+": 1.5, "D0": 1.0,
    "F": 0
    }

# 2. 분야-과목-가중치 데이터
fields = {
    "1": {
        "name": "생산/제조/제품설계",
        "subjects": {
            "일반화학": 2,
            "기초미적분학": 2,
            "스마트팩토리개론": 1,
            "이산수학": 5,
            "작업설계및측정": 5,
            "객체지향프로그래밍": 5,
            "인지공학개론": 4,
            "실험계획법": 5,
            "인간공학": 5,
            "생산운영관리": 4
        }
    },
    "2": {
        "name": "물류/유통",
        "subjects": {
            "기초통계학": 2,
            "4차산업혁명 이해": 1,
            "선형대수": 5,
            "기초경영과학": 5,
            "경영과학응용": 5,
            "공급사슬관리": 4
        }
    },
    "3": {
        "name": "품질/안전",
        "subjects": {
            "물리학및실험1": 2,
            "공업통계": 5,
            "품질경영개론": 5,
            "공업통계응용": 4,
            "스마트팩토리맵": 5
        }
    },
    "4": {
        "name": "전략/기획",
        "subjects": {
            "기초물리학": 2,
            "발표와토론": 1,
            "산업경영공학개론": 5,
            "파이썬프로그래밍": 5,
            "이산수학(전략)": 5,
            "기술경영개론": 3,
            "데이터베이스": 5,
            "경영정보시스템": 3
        }
    }
}

# ---------------------------------------------
# 3. 사용자에게 분야 선택 받기
# ---------------------------------------------
print("=== 진출 분야 선택 ===")
print("1. 생산/제조/제품설계")
print("2. 물류/유통")
print("3. 품질/안전")
print("4. 전략/기획")

choice = input("원하는 분야 번호를 입력하세요: ").strip()
while choice not in fields:
    choice = input("잘못된 입력입니다. 번호를 다시 입력하세요: ").strip()

selected_field = fields[choice]
print(f"\n👉 선택한 분야: {selected_field['name']}")

# ---------------------------------------------
# 4. 선택 분야의 과목만 학점 입력
# ---------------------------------------------
user_grades = {}
print("\n=== 학점 입력 (A+, A0, B+, B0, C+, C0, D+, D0, F) ===")

for subject in selected_field["subjects"]:
    grade = input(f"{subject} 학점 입력: ").strip().upper()
    while grade not in grade_to_score:
        grade = input("잘못된 학점입니다. 다시 입력하세요: ").strip().upper()
    user_grades[subject] = grade_to_score[grade]

# ---------------------------------------------
# 5. 점수 계산
# ---------------------------------------------
total_score = 0
for subject, weight in selected_field["subjects"].items():
    total_score += user_grades[subject] * weight

print("\n==============================")
print(f"📌 분야: {selected_field['name']}")
print(f"📊 총점: {total_score:.2f}")
print("==============================\n")

# ---------------------------------------------
# 6. 추천 정보 출력
# ---------------------------------------------
recommendations = {
    "생산/제조/제품설계": {
        "자격증": ["품질경영기사", "생산자동화산업기사", "산업안전기사", "ERP정보관리사(생산)"],
        "취업 분야": ["생산기술 엔지니어", "제조 공정 엔지니어", "자동화 엔지니어", "제품설계"]
    },
    "물류/유통": {
        "자격증": ["물류관리사", "유통관리사", "ERP 물류", "SCM 전문가"],
        "취업 분야": ["SCM 분석", "물류 운영관리", "재고관리", "운송계획"]
    },
    "품질/안전": {
        "자격증": ["품질경영기사", "신뢰성기사", "산업안전기사"],
        "취업 분야": ["품질보증(QA)", "품질관리(QC)", "안전관리자", "공정 품질 엔지니어"]
    },
    "전략/기획": {
        "자격증": ["ADsP", "빅데이터 분석기사", "경영지도사", "PMP"],
        "취업 분야": ["전략기획", "경영지원", "데이터기획", "서비스기획"]
    }
}

field_name = selected_field["name"]

print("🎯 추천 자격증:")
for cert in recommendations[field_name]["자격증"]:
    print(" -", cert)

print("\n💼 추천 취업 분야:")
for job in recommendations[field_name]["취업 분야"]:
    print(" -", job)

# ---------------------------------------------
# 7. 결과를 엑셀 파일로 저장
# ---------------------------------------------

wb = Workbook()
ws = wb.active
ws.title = "진출분야 결과"

# 제목
ws.append(["분야", "총점", "추천 자격증", "추천 취업 분야"])

# 데이터 입력
ws.append([
    field_name,
    total_score,
    ", ".join(recommendations[field_name]["자격증"]),
    ", ".join(recommendations[field_name]["취업 분야"])
])

# 파일 저장
excel_filename = "result.xlsx"
wb.save(excel_filename)

print(f"\n📁 결과 파일이 '{excel_filename}' 이름으로 저장되었습니다!")
```
